```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.matlib
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.image as mpimg
import cv2
```

### Function converting images into column vector containing 1 and -1

```python
def jpg2bin(image, thresh):
    image_gray = cv2.imread(image, cv2.IMREAD_GRAYSCALE)
    image_bw = cv2.threshold(image_gray, thresh, 255, cv2.THRESH_BINARY)[1]
    image_bin=image_bw/255.0
    img_arr=image_bin*2-1;
    img_vec = np.reshape(img_arr, (1,40000))
    return img_vec
```

### Function for adding random noise

```python
def noisy(image_name, mag):
    image = cv2.imread(image_name,1)
    noised_image = image + mag * image.std() * np.random.random(image.shape)
    RGB_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    saved_image = cv2.imwrite("noisy.jpg", noised_image)
    noisy = mpimg.imread('noisy.jpg')
    return 'noisy.jpg'
```

### Function computing inner product of received image and each database image

```python
def image_inner(r, s1, s2, s3, s4, s5, s6, s7, s8):
    plt.figure(figsize=(14, 8))
    plt.suptitle('<Recived Image, Database Images>', fontsize=24)
    ax1=plt.subplot(2, 4, 1)
    ax1.imshow(brian)
    ax1.set_title('%d' % np.inner(r,s1), fontsize=28, color='red')
    ax2=plt.subplot(2, 4, 2)
    ax2.imshow(iggy)
    ax2.set_title('%d' % np.inner(r,s2), fontsize=28, color='red')
    ax3=plt.subplot(2, 4, 3)
    ax3.imshow(santa)
    ax3.set_title('%d' % np.inner(r,s3), fontsize=28, color='red')
    ax4=plt.subplot(2, 4, 4)
    ax4.imshow(sadaharu)
    ax4.set_title('%d' % np.inner(r,s4), fontsize=28, color='red')
    ax5=plt.subplot(2, 4, 5)
    ax5.imshow(droopy)
    ax5.set_title('%d' % np.inner(r,s5), fontsize=28, color='red')
    ax6=plt.subplot(2, 4, 6)
    ax6.imshow(marmaduke)
    ax6.set_title('%d' % np.inner(r,s6), fontsize=28, color='red')
    ax7=plt.subplot(2, 4, 7)
    ax7.imshow(ein)
    ax7.set_title('%d' % np.inner(r,s7), fontsize=28, color='red')
    ax8=plt.subplot(2, 4, 8)
    ax8.imshow(snoopy)
    ax8.set_title('%d' % np.inner(r,s8), fontsize=28, color='red')
    return ()
```

### Plotting database images

```python
brian=mpimg.imread('brian.jpg')
s1=jpg2bin('brian.jpg', 128)
iggy=mpimg.imread('iggy.jpg')
s2=jpg2bin('iggy.jpg', 128)
santa=mpimg.imread('santa.jpg')
s3=jpg2bin('santa.jpg', 128)
sadaharu=mpimg.imread('sadaharu.jpg')
s4=jpg2bin('sadaharu.jpg', 128)
droopy=mpimg.imread('droopy.jpg')
s5=jpg2bin('droopy.jpg', 128)
marmaduke=mpimg.imread('marmaduke.jpg')
s6=jpg2bin('marmaduke.jpg', 128)
ein=mpimg.imread('ein.jpg')
s7=jpg2bin('ein.jpg', 128)
snoopy=mpimg.imread('snoopy.jpg')
s8=jpg2bin('snoopy.jpg', 128)

plt.figure(figsize=(14, 7))
plt.suptitle('Database Images', fontsize=24)
ax1=plt.subplot(2, 4, 1)
ax1.imshow(brian)
ax2=plt.subplot(2, 4, 2)
ax2.imshow(iggy)
ax3=plt.subplot(2, 4, 3)
ax3.imshow(santa)
ax4=plt.subplot(2, 4, 4)
ax4.imshow(sadaharu)
ax5=plt.subplot(2, 4, 5)
ax5.imshow(droopy)
ax6=plt.subplot(2, 4, 6)
ax6.imshow(marmaduke)
ax7=plt.subplot(2, 4, 7)
ax7.imshow(ein)
ax8=plt.subplot(2, 4, 8)
ax8.imshow(snoopy)


```
