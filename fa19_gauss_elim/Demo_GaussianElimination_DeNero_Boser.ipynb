```python
import matplotlib
from pylab import imread, imshow
from IPython.display import Image, display
import numpy as np
import numpy.matlib
import matplotlib.pyplot as plt

%matplotlib inline

instr1 = imread("denero.jpg")
instr2 = imread("boser.jpg")

instr1 = instr1[:,:,0]
instr2 = instr2[:,:,0]

numrows = instr1.shape[0]
numcols = instr2.shape[1]
```

```python
factor1a = .61
factor2a = 1 - factor1a
mix1 = factor1a*instr1 + factor2a*instr2

factor1b = .16
factor2b = 1 - factor1b
mix2 = factor1b*instr1 + factor2b*instr2
```

```python
plt.imshow(mix1, cmap = 'gray')
```

```python
plt.imshow(mix2, cmap = 'gray')
```

```python
ratios = np.array([[factor1a, factor2a], [factor1b, factor2b]])

inst1vec = np.reshape(mix1, (1,numrows*numcols))
inst2vec = np.reshape(mix2, (1,numrows*numcols))

bvec = np.vstack((inst1vec, inst2vec))
bvec.shape
```

```python
answer  = np.linalg.solve(ratios, bvec)
```

```python
answer.shape
```

```python
answer1 = answer[0,:]
answer2 = answer[1,:]
```

```python
answer11 = np.reshape(answer1,(numrows,numcols))
answer22 = np.reshape(answer2,(numrows,numcols))
```

```python
plt.imshow(answer11, cmap ='gray')
```

```python
inst1vec = np.reshape(mix1, (1,numrows*numcols))
inst2vec = np.reshape(mix1, (1,numrows*numcols))

bvec = np.vstack((inst1vec, inst2vec))

ratios = np.array([[factor1a, factor1b], [2*factor1b, 2*factor2b]])

answer  = np.linalg.solve(ratios, bvec)

answer1 = answer[0,:]
answer2 = answer[1,:]

answer11 = np.reshape(answer1,(numrows,numcols))
answer22 = np.reshape(answer2,(numrows,numcols))
```
